#데이터셋 확인

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/winequality-white.csv", sep=';')

df
```

#결측치 확인

```python
df['quality'].value_counts()
```

#데이터 나누기

```python
#입력과 출력 나누기
X = df.drop(columns=['quality'])
Y = df['quality']

X
```

```python
Y
```

#결측치를 0부터 인코딩하기

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
encoder = LabelEncoder()

# Y 데이터를 인코딩하여 Y_encoded 변수에 저장
Y_encoded = encoder.fit_transform(Y)

# 변환된 라벨을 Y에 할당하여 업데이트
Y = Y_encoded

# 업데이트된 Y의 상위 5개 값 확인
print("업데이트된 Y 데이터의 상위 5개:\n", Y[:5])

# 업데이트된 Y의 고유한 값과 개수 확인
print("\n업데이트된 Y의 고유한 값들:", pd.Series(Y).unique())
print("\n업데이트된 Y의 값별 개수:\n", pd.Series(Y).value_counts().sort_index())
```

```python
Y
```

#train과 Test로 나누기

```python
#train과 test 나누기
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state=7 ,stratify=Y)

#X 확인
X
```

```python
Y
```

#훈련모델 설정 및 훈련

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

#DT
dt = tree.DecisionTreeClassifier();
dt.fit(X_train,Y_train)

#RF
rf = RandomForestClassifier();
rf.fit(X_train,Y_train)

#LR
lr = LogisticRegression();
lr.fit(X_train,Y_train)

#KNN
knn = KNeighborsClassifier();
knn.fit(X_train,Y_train)
```

#정확도 분석 및 기울기 및 절편구하기

```python
from sklearn.metrics import accuracy_score

#DT
Y_pred_dt = dt.predict(X_test)
print("DT 정확도:{}".format(accuracy_score(Y_test,Y_pred_dt)))

#RF
Y_pred_rf = rf.predict(X_test)
print("RF 정확도:{}".format(accuracy_score(Y_test,Y_pred_rf)))

#LR
Y_pred_lr = lr.predict(X_test)
print("LR 정확도:{}".format(accuracy_score(Y_test,Y_pred_lr)))
print("LR 계수:", lr.coef_)
print("LR 절편:", lr.intercept_)

#KNN
Y_pred_knn = knn.predict(X_test)
print("KNN 정확도:{}".format(accuracy_score(Y_test,Y_pred_knn)))
```
